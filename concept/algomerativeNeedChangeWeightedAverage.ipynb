```
import numpy as np
def Euclidian_Dist(x,y):
  return np.linalg.norm(x-y) #write your code here

def Dist_mat(X):
  #write your code here
  dist_mat=np.zeros((len(X),len(X)))
  for i in range(len(X)):
    for j in range(len(X)):
      dist_mat[i,j]=round(Euclidian_Dist(X[i],X[j]),2) if i!=j else np.inf
  return dist_mat

def combine(X,D):
  #write your code here
  a,b = np.unravel_index(np.argmin(D, axis=None), D.shape)
  a,b=(b,a) if b<a else (a,b)
  print("Vector of X to be combined : " + str(a+1) + "," + str(b+1))
  X[a]=(X[a]+X[b])/2
  newX=np.delete(X,b,0)
  D[a,:]=(D[a,:]+D[b,:])/2
  D[:,a]=(D[:,a]+D[:,b])/2
  D=np.delete(D,b,0)
  D=np.delete(D,b,1)
  return newX,a,b,D

X=np.array([[1,1],[2,1],[5,4],[6,5],[6.5,6]])
print(X.T)
#write your code here
n=1 #number_of_cluster
newX=X
label=[[i] for i in range(len(X))]
D=Dist_mat(newX)
while newX.shape[0]>n:
  print(D)
  newX,a,b,D=combine(newX,D)
  label[a]+=label[b]
  label.pop(b)
  print()
  print(newX.T)
print(label)
## velidate from inbuilt Dendogram

import plotly.figure_factory as ff

lab=np.linspace(1,X.shape[0],X.shape[0])
fig = ff.create_dendrogram(X, labels=lab)
fig.update_layout(width=800, height=300)
fig.show()
```
